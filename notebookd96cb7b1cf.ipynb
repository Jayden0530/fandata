```python
#머신러닝으로 돌릴거여서 1차원으로 축소해야한다.
#for 문으로 돌린다.

import warnings
warnings.filterwarnings("ignore") 

from tqdm import tqdm
train_list = []

for path in tqdm(train["SAMPLE_PATH"]):
    data,sr=librosa.load(path,res_type='kaiser_fast',duration=4) #best 로 하는게 좋음 그러나 별차이 없음.
    mfcc=librosa.feature.mfcc(data,n_mfcc=128) 
    mfcc_median = np.median(mfcc,axis=1)
#     mfcc_std=np.std(mfcc,axis=1)
#     mfcc1=pd.DataFrame(mfcc_median)
#     mfcc2=pd.DataFrame(mfcc_std)
    train_list.append(mfcc_median)
```

```python
from tqdm import tqdm
test_list = []

for path in tqdm(test["SAMPLE_PATH"]):
    data,sr=librosa.load(path,res_type='kaiser_fast',duration=4) #best 로 하는게 좋음 그러나 별차이 없음.
    mfcc=librosa.feature.mfcc(data,n_mfcc=128) 
    mfcc_median=np.median(mfcc,axis=1)
    test_list.append(mfcc_median)    
```

```python
pd.DataFrame(train_list)
```

```python
#비지도학습 모델링

from pycaret.anomaly import *
ano = setup(data=train_list,normalize=True,session_id=242)
```

```python
models()
```

```python
sub=pd.read_csv("/kaggle/input/dacon-fan-data/sample_submission.csv")
sub
```

```python
result=[]
ensangble_list=np.zeros(len(test))
for model_name in tqdm(models()["Name"].index[:-2]):
    model=create_model(model_name,fraction=0.28)
    model_fit=assign_model(model)
    prediction = predict_model(model,test_list)["Anomaly"]
    result.append(prediction.tolist())
    ensangble_list+= prediction
```

```python
sub["LABEL"] = ensangble_list > 7 #잴 점수가 잘나왔음.
```
